```python
import pandas as pd
import random
import numpy
import datetime
from datetime import timedelta
```

# Help Functions

```python


# returns a random time

def random_time():
    
    # generate random number scaled to number of seconds in a day
    # (24*60*60) = 86,400
    
    rtime = int(random.random()*86400)
    
    hours   = int(rtime/3600)
    minutes = int((rtime - hours*3600)/60)
    seconds = rtime - hours*3600 - minutes*60
    
    time_string = '%02d:%02d:%02d' % (hours, minutes, seconds)
    return time_string



#returns a random date
def random_date(start, end):
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = random.randrange(0,int_delta)
    data =start + timedelta(seconds=random_second)
    return data
    

#generate_random_address
def generate_random_address():
  street_names = ['Main', '2nd', '1st', '4th', '5th', 'Park', '6th', '7th', 'Maple', 'Pine', 'Washington', '8th', 'Cedar', 'Elm', 'Walnut', '9th', '10th', 'Lake', 'Sunset', 'Lincoln', 'Jackson', 'Church', 'River', '11th', 'Willow', 'Jefferson', 'Center', '12th', 'North', 'Lakeview', 'Ridge', 'Hickory', 'Adams', 'Cherry', 'Highland', 'Johnson', 'South', 'Dogwood', 'West', 'Chestnut', '13th', 'Spruce', '14th', 'Wilson', 'Meadow', 'Forest', 'Hill', 'Madison']
  cities = ['San Francisco', 'Boston', 'New York City', 'Austin', 'Dallas', 'Atlanta', 'Portland', 'Portland', 'Los Angeles', 'Seattle']
  weights = [9,4,5,2,3,3,2,0.5,6,3]
  zips = ['94016', '02215', '10001', '73301', '75001', '30301', '97035', '04101', '90001', '98101']
  state = ['CA', 'MA', 'NY', 'TX', 'TX', 'GA', 'OR', 'ME', 'CA', 'WA']
  
  street = random.choice(street_names)
  index = random.choices(range(len(cities)), weights=weights)[0]

  return f"{random.randint(1,999)} {street} St, {cities[index]}, {state[index]} {zips[index]}"

def cria_lista(id, mes, data_inicio, data_final):
    
    # list comprehension to select the product list
    lista_produtos = [produto for produto in products]
    # List Comprehension to select product weights
    pesos = [products[produto][1] for produto in products]
    
    # Uses the Gaussian distribution to randomly generate the amount of sales in the month
    if mes <=10:
            orders_amount = int(numpy.random.normal(loc=12000, scale=4000))
    if mes == 11:
            orders_amount = int(numpy.random.normal(loc=20000, scale=3000))
    if mes ==12:
            orders_amount = int(numpy.random.normal(loc=26000, scale=3000))
            
    order_id = 1
    i = 1        
    while orders_amount>0:
        
        # Using random.choices to return a random product, influenced by the weight of the product. Random choices returns a list
        produto = random.choices(lista_produtos, weights = pesos)[0]
        # Select product price
        preco = products[produto][0]
        # Generating the order quantity, using numpy.geometric
        quantity_ordered = numpy.random.geometric(p=1.0-(1.0/preco),size=1)[0]
        # Generating random date
        data = random_date(data_inicio, data_final)
        # Generating random address
        endereco = generate_random_address()
        # Assigns the values ​​to the dataframe
        df.loc[i] = [order_id, produto, quantity_ordered, preco, data, endereco]
        i +=1
        # Randomly adding products, usually purchased together for the same purchase
        if produto == 'Iphone':
            if random.random() <0.15:
                df.loc[i] = [order_id, 'Lightning Charging Cable', quantity_ordered, products['Lightning Charging Cable'][0], data, endereco]
                i+=1
            if random.random() <0.05:
                df.loc[i] = [order_id, 'Apple Airpoods Headphones', quantity_ordered, products['Apple Airpoods Headphones'][0], data, endereco]
                i+=1
            if random.random() <0.05:
                df.loc[i] = [order_id, 'Wired Headphones', quantity_ordered, products['Wired Headphones'][0], data, endereco]
                i+=1
        elif produto == 'Google Phone':
            if random.random() <0.15:
                df.loc[i] = [order_id, 'Wired Headphones', quantity_ordered, products['Wired Headphones'][0], data, endereco]
                i+=1
            if random.random() <0.05:
                df.loc[i] = [order_id, 'Apple Airpoods Headphones', quantity_ordered, products['Apple Airpoods Headphones'][0], data, endereco]
                i+=1
            if random.random() <0.30:
                df.loc[i] = [order_id, 'USB-C Charging Cable', quantity_ordered, products['USB-C Charging Cable'][0], data, endereco]
                i+=1
        elif produto == '34in Ultrawide Monitor' or produto == '27in 4k Gaming Monitor'or produto=='27in FHD Monitor':
            if random.random() <0.45:
                df.loc[i] = [order_id, 'AA Batteries (4-pack)', quantity_ordered, products['AA Batteries (4-pack)'][0], data, endereco]
                i+=1
            
                
        if random.random()<=0.20:
            product2=  random.choices(lista_produtos, weights = pesos)[0]
            df.loc[i] = [order_id, product2, quantity_ordered, products[product2][0], data, endereco]
            i+=1
        
            if random.random()<=0.02:
                product3=  random.choices(lista_produtos, weights = pesos)[0]
                df.loc[i] = [order_id, product3, quantity_ordered, products[product3][0], data, endereco]
                i+=1
            
            elif random.random()<=0.01:
                product4=  random.choices(lista_produtos, weights = pesos)[0]
                df.loc[i] = [order_id, product4, quantity_ordered, products[product4][0], data, endereco]
                i+=1
        order_id +=1
        orders_amount -=1
```
